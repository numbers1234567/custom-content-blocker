```python
import time
response = requests.get(f"https://newsdata.io/api/1/latest?apikey={api_key}&language=en&category=politics&prioritydomain=top&size=10")
content = json.loads(response.content)
pubdates = []

while len(content["results"]) > 0:
    for item in content["results"]:
        pubdates.append(item["pubDate"] + " " + item["pubDateTZ"])
    response = requests.get(f"https://newsdata.io/api/1/latest?apikey={api_key}&language=en&category=politics&prioritydomain=top&size=10&page={content['nextPage']}")
    content = json.loads(response.content)
    time.sleep(1)
```

Using the /latest API, news from the last 48 hours can be retrieved, but also delayed by 12 hours. 

It appears publish dates are sorted by most-recent-first, which makes working with the API a bit easier. 

In the above test, the API indicates 2429 results available. However, there is a limit of 200 requests per day meaning only 2000 of these results can actually be processed on any day. This is not an issue because the internal database will store some news stories, so there is no reason to read all the results given that the database already contains the earlier ones.

The actual content of the news story is not available: just the title, image, and description. The other APIs that I was evaluating also did not have the content available.
